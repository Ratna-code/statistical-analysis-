**Chi-Square Test**

The test is applied when you have two categorical variables from a single population. It is used to determine whether there is a significant association between the two variables.

- null hypothesis: There is a significant association between sex and smoker
- Alternate  hypothesis: There is no  significant association between sex and smoker

```python
import pandas as pd
from scipy import stats
import seaborn as sns
# Load the tips dataset
tips = sns.load_dataset("tips")
# Create a contingency table of sex vs. smokers
contingency_table = pd.crosstab(tips["sex"], tips["smoker"],margins =True)
contingency_table
```

```python

# Perform a chi-square test
chi2, p, dof, expected = stats.chi2_contingency(contingency_table)

print(f"Chi-square statistic: {chi2:.4f}")
print(f"P-value: {p:.4f}")
print(f"Degrees of freedom: {dof}")
print(f"Expected frequencies:\n{expected}")

# Set the alpha level
alpha = 0.05
# Interpret the results
if p < alpha:
    print("Reject the null hypothesis: There is a significant association between sex and smoker.")
else:
    print("Fail to reject the null hypothesis: There is no significant association between sex and smoker.")
```

```python

import pandas as pd
from scipy import stats
import seaborn as sns

# Load the Titanic dataset
titanic = sns.load_dataset("titanic")

# Create a contingency table of survived vs. sex
contingency_table = pd.crosstab(titanic["survived"], titanic["sex"], margins=True)

# Perform a chi-square test
chi2, p, dof, expected = stats.chi2_contingency(contingency_table)

# Print the results
print(f"Degrees of freedom: {dof}")
print(f"Expected frequencies:\n{expected}")
print(f"Chi-square statistic: {chi2:.4f}")
print(f"P-value: {p:.4f}")


# Set the alpha level
alpha = 0.05

# Interpret the results
if p < alpha:
    print("Reject the null hypothesis: There is a significant association between survived and sex.")
else:
    print("Fail to reject the null hypothesis: There is no significant association between survived and sex.")
```
